```python
import re 
def clean_benef(benef):
    if pd.isnull(benef):
        return np.nan
    if isinstance(benef, (str)): 
        return benef.split(',')
    else : 
        return np.nan

all_benefL=all_benef.apply(clean_benef)
```

```python
all_benefL.dropna()
```

```python
Liste_COBenef = all_benefL[0]
clean_benef_liste = []

for liste_benef in all_benefL.dropna():
    if isinstance(liste_benef, list):  # Vérifie si l'élément est une liste
        cleaned_list = []
        for benef in liste_benef:
            if benef in Liste_COBenef:
                cleaned_list.append(benef)
        clean_benef_liste.append(cleaned_list)
    else:
        # Si l'élément n'est pas une liste, ajoute-le tel quel à la liste nettoyée
        clean_benef_liste.append(liste_benef)
clean_benef_liste
```

```python
#print(clean_benef_liste)

#family=family.value_counts()
from collections import Counter

# Concaténer toutes les listes nettoyées en une seule liste
flattened_list = [item for sublist in clean_benef_liste for item in sublist]

# Utiliser Counter pour compter les occurrences de chaque élément dans la liste
occurrences = Counter(flattened_list)

# Afficher les résultats
for element, count in occurrences.items():
    print(f"{element}: {count} occurrence(s)")
```

```python
elements = list(occurrences.keys())
counts = list(occurrences.values())
# Création du plot
plt.figure(figsize=(10, 6))
plt.bar(elements,counts)
plt.title('Nombre d\'occurrences par Co benefices')
plt.xlabel('Co benefice')
plt.ylabel('Nombre d\'occurrences')
plt.xticks(rotation=45)  # Rotation des étiquettes de l'axe des x pour une meilleure lisibilité
plt.grid(axis='y')  # Afficher une grille uniquement sur l'axe des y

# Sauvegarde du plot dans le dossier courant
plt.savefig('Top10_benef.png')

# Affichage du plot
plt.show()
```
